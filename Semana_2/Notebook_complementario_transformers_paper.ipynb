```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_angles(pos, i, d_model):
  angle_rates = 1 / np.power(10000, (2 * (i//2)) / np.float32(d_model))
  return pos * angle_rates

def positional_encoding(position, d_model):
  angle_rads = get_angles(np.arange(position)[:, np.newaxis],
                          np.arange(d_model)[np.newaxis, :],
                          d_model)

  # Seno a los indices pares 2i
  angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])

  # Coseno a los indices impares 2i+1
  angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])

  pos_encoding = angle_rads[np.newaxis, ...]

  return pos_encoding
```

```python
plt.rcParams['figure.dpi'] = 150

tokens = 4096
dimensions = 12288


pos_encoding = positional_encoding(tokens, dimensions)
print (pos_encoding.shape)
plt.figure(figsize=(12,8))
plt.pcolormesh(pos_encoding[0], cmap='Blues')
plt.xlabel('Dimensiones del Embedding\n', fontsize=9)
plt.xlim((0, dimensions))
plt.ylim((tokens,0))
plt.ylabel('\nPosition del token en el input', fontsize=9)
plt.colorbar()
plt.title('\nCodificacion Posicional en los transformers - Incidencia de la posicion en los embeddings\n', fontsize=10)
plt.show()
```
